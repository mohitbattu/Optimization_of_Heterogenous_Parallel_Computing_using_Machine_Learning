```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import time
```

```python
#Convert to dictionary
leaf_size = [1]
n_neighbors = [2]
p=[1]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import KFold 
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)
knn = KNeighborsClassifier()
clf_knn = GridSearchCV(knn, hyperparameters, cv=6)
kn_tr_tim=[]
kn_pr_tim=[]
scores_knn=[]
scores_ktr=[]
for train_index, test_index in kf.split(data,target):
    print("Train:", train_index, "Validation:",test_index)
    X_train, X_test = data[train_index], data[test_index] 
    y_train, y_test = target[train_index], target[test_index]
    training_start = time.perf_counter()
    best_model=clf_knn.fit(X_train, y_train)
    training_end = time.perf_counter()
    Knn_train=training_end-training_start
    kn_tr_tim.append(Knn_train)
    prediction_start = time.perf_counter()
    knn_pred = clf_knn.predict(X_test)
    prediction_end = time.perf_counter()
    Knn_pred=prediction_end-prediction_start
    kn_pr_tim.append(Knn_pred)
    scores_knn.append(clf_knn.score(X_test,y_test))
    scores_ktr.append(clf_knn.score(X_train,y_train))
knn_train_time = np.mean(kn_tr_tim)
knn_prediction_time = np.mean(kn_pr_tim)
acc_knn=np.mean(scores_knn)
acc_tr_knn=np.mean(scores_ktr)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold 
kf =StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rfc = RandomForestClassifier(n_estimators=100)

rf_tr_tim=[]
rf_pr_tim=[]
scores_rfc=[]
scores_rtr=[]
for train_index, test_index in kf.split(data,target):
    print("Train:", train_index, "Validation:",test_index)
    X_train, X_test = data[train_index], data[test_index] 
    y_train, y_test = target[train_index], target[test_index]
    training_start = time.perf_counter()
    best_model=rfc.fit(X_train, y_train)
    training_end = time.perf_counter()
    rfc_train=training_end-training_start
    rf_tr_tim.append(rfc_train)
    prediction_start = time.perf_counter()
    rf_pred = rfc.predict(X_test)
    prediction_end = time.perf_counter()
    rfc_pred=prediction_end-prediction_start
    rf_pr_tim.append(rfc_pred)
    scores_rfc.append(rfc.score(X_test,y_test))
    scores_rtr.append(rfc.score(X_train,y_train))

rfc_train_time = np.mean(rf_tr_tim)
rfc_prediction_time = np.mean(rf_pr_tim)
acc_rfc=np.mean(scores_rfc)
acc_tr_rfc=np.mean(scores_rtr)
```
